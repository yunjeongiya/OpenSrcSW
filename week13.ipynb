```python
#실습1 생일 기온 시각화
import matplotlib.pyplot as plt
import csv
f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
maxlist=[]
minlist=[]
for row in data :
    if row[-1] != '':
        if row[0].split('-')[1] == '08' and row[0].split('-')[2]=='13':
            maxlist.append(float(row[-1]))
            minlist.append(float(row[-2]))

plt.plot(maxlist, color='deeppink')
plt.plot(minlist, color='deepskyblue')
plt.show()
f.close()
```

```python
#실습 2 최고 기온 시각화
import matplotlib.pyplot as plt
import csv
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
jan=[]
aug=[]
for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month == '01':
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))

plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
f.close()
```
