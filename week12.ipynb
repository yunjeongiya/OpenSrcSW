```python
#실습1 서울의 일교차

import csv
f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',') #','가 기본값이므로 그냥 csv.reader(f) 만 해도 가능
header = next(data)

max_temp = -999
max_date = ''

min_temp = 999
min_date = ''

max_gap_temp = 0
max_gap_date = ''

for row in data:
    if row[-1] == '':
        row[-1] = -999
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
        
    if row[-2] == '':
        row[-2] = 999
    row[-2] = float(row[-2])
    if min_temp > row[-2]:
        min_date = row[0]
        min_temp= row[-2]
        
    if row[-1] - row[-2] > max_gap_temp:
        max_gap_date = row[0]
        max_gap_temp = row[-1] - row[-2]

f.close()

print("일교차가 가장 큰 날\n", max_gap_date)
print("최저기온이 가장 낮은 날\n", min_date)
print("최고기온이 가장 높은 날\n", max_date)
```
